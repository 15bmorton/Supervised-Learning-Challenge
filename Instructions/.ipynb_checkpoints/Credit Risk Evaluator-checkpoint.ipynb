```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df
```

```python
train_df.count()
```

```python
test_df.count()
```

```python
YN_dict = {'Y':1, 'N': 0}
train_df1 = train_df.replace({'hardship_flag': YN_dict, 'debt_settlement_flag': YN_dict})
Home_dict = {'ANY':0, 'RENT':1, 'MORTGAGE': 2, 'OWN':3}
train_df2 = train_df1.replace({'home_ownership': Home_dict})
verif_dict = {'Not Verified':0, 'Source Verified': 1, 'Verified': 1}
train_df3 = train_df2.replace({'verification_status': verif_dict})
loanstat_dict = {'low_risk':1, 'high_risk':0}
train_df4 = train_df3.replace({'loan_status': loanstat_dict})
init_list_stat_dict = {'w':0, 'f':1}
train_df5 = train_df4.replace({'initial_list_status': init_list_stat_dict})
app_type_dict= {'Individual':1, 'Joint App': 0}
train_df6 = train_df5.replace({'application_type': app_type_dict})
```

```python
train_revised = train_df6.drop(['index', 'pymnt_plan', 'Unnamed: 0'], axis =1)
X_train = train_revised.drop('loan_status', axis =1)
y_train = train_revised['loan_status'].values
train_revised
```

```python
test_df1 = test_df.replace({'hardship_flag': YN_dict, 'debt_settlement_flag': YN_dict})
test_df2 = test_df1.replace({'home_ownership': Home_dict})
test_df3 = test_df2.replace({'verification_status': verif_dict})
test_df4 = test_df3.replace({'loan_status': loanstat_dict})
test_df5 = test_df4.replace({'initial_list_status': init_list_stat_dict})
test_df6 = test_df5.replace({'application_type': app_type_dict})
```

```python
test_revised = test_df6.drop(['index', 'pymnt_plan', 'Unnamed: 0'], axis =1)
X_test = test_revised.drop('loan_status', axis =1)
y_test = test_revised['loan_status'].values
test_revised
```

```python
#The Random Forest Classifier model will perform better because 
#it is made to handle unbalanced data and is very versatile.
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter=200)
classifier.fit(X_train, y_train)
print(f"Train Data Score: {classifier.score(X_train, y_train)}")
print(f"Test Data Score: {classifier.score(X_test, y_test)}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=1, n_estimators=500). fit(X_train, y_train)
```

```python
print(f"Train Data Score: {rfc.score(X_train, y_train)}")
print(f"Test Data Score: {rfc.score(X_test, y_test)}")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
classifier.fit(X_train_scaled, y_train)
```

```python
print(f"Train Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Test Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
print(f"Test Score: {rfc.score(X_test_scaled, y_test)}")
```
